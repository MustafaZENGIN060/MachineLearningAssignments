### Deal with Outliers

#### *ii. Drop Outliers*

```python
def out_lüger(df_in, column):
    q1=df_in[column].quantile(0.25)
    q3=df_in[column].quantile(0.75)
    ıqr=q3-q1
    low_lim=q1-3*ıqr
    high_lim=q3+3*ıqr
    df_out_ind=df_in[~((df_in[column]<low_lim) | (df_in[column]>high_lim))]
    return df_out_ind
```

```python
out_lüger(df,"Horizontal_Distance_To_Hydrology")
```

- As second, we can produce&transform an additional column which contains **average** of Horizantal Distances to Hydrology and Roadways.

- Third, I decided to transform a new column which contains **average** of `Elevation` and ``Vertical_Distance_To_Hydrology`` columns.
